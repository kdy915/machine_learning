```python
#딕셔너리를 이용해서 데이터프레임 만들기
df=pd.DataFrame(data={"col1":[1,2],"col2":[3,4]})
df
```

```python
df=pd.DataFrame(data=[{"col1":1,"col2":3},{"col1":2,"col2":4}])
df
```

```python
#데이터가 컬럼수만큼 채워지지 않으면 NaN값 지정한다.
df=pd.DataFrame(data=[{"col1":1,"col2":3},{"col1":2,"col2":4},{"col2":5}])
df
```

```python
#리스트를 이용해서 데이터 프레임 만들기
L1=[1,2,3,4,5]
L2=[6,7,8,9,10]
df=pd.DataFrame(data={"col1":L1,"col2":L2})
df
```

```python
# numpy c_[]를 이용하면 열 단위로 데이터 생성
import numpy as np
df=pd.DataFrame(np.c_[L1,L2],columns=["col1","col2"])
df
```

```python
import numpy as np
L1=[[1,2,3,4,5]]
L2=[[6,7,8,9,10]]
df=pd.DataFrame(np.r_[L1,L2])
df
```

```python
#iris 데이터 scikit-learn 패키지에서 가져오기
import pandas as pd
from sklearn import datasets
```

```python
iris_dic=datasets.load_iris()
iris_dic
```

```python
# 필요한 데이터들을 가져온다
x=pd.DataFrame(iris_dic.data,columns=iris_dic['feature_names'])
y=pd.DataFrame(iris_dic['target_names'][iris_dic['target']],columns=['species'])
print(x.shape)
print(y.shape)
```

```python
# 아이리스 데이터 상위 5개만 출력 x,y데이터 axis로 결합
iris=pd.concat([x,y],axis=1)
iris.head()
```

```python
# 데이터베이스 열이름을 변경합니다.
iris.columns=["sepal_length","sepal_width","petal_length","petal_width","species"]
```

```python
#seaborn에서 아이리스 데이터를 가져옵니다,
# load_dataset에는 여러가지 데이터셋들이 있다.
import seaborn as sns
iris=sns.load_dataset("iris")
iris.head()
```

```python
#데이터프레임 파일을 csv파일에 저장한다.
iris.to_csv("iris.csv",sep=',', index=False)
```

```python
#gzip 압축을 이용해서 압축합니다
# index는 값이 True,False이면 칼럼값 입력, compression은 압축 방법입니다.
iris.to_csv("iris.csv.gz",sep=',', index=False, compression='infer')
```
