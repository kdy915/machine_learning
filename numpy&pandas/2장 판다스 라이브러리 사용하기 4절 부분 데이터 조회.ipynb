```python
#조건을 통해서 조회하기
iris.loc[iris['species']=='versicolor'].head()
```

```python
iris.loc[iris["species"]=="versicolor",["sepal_length","sepal_width"]].head()
```

```python
#versicolor 종이며 sepal_length가 6.5볻 큰 데이터 조회
# & 대신 and를 사용하면 에러가 발생한다
iris.loc[(iris['species']=='versicolor')& (iris['sepal_length']>6.5)].head()
```

```python
import pandas as pd
df=pd.DataFrame({'X':[1,2,3],'Y':[6,7,8]},index=['A','B','C'])
```

```python
# 다음과 같이 행을 뽑고 열을 찾아서 수정하면 수정이 되지 않는다
df.loc[df['Y']%2==0].loc[:,'X']=[10,30]
```

```python
# 열을 찾고 그다음 행을 찾아서 수정해야 한다.
df.loc[:,'X'].loc[df['Y']%2==0]=[10,30]
df
```

```python
# 행과 열을 한번에 조회해서 수정 가능
df.loc[df['Y']%2==0,'X']=[10,30]
```

### 데이터 삭제 및 추가

```python
# 0행을 삭제합니다. 원본데이터는 inplace=True를 해줘야 반영됩니다.
iris.drop(0)
```

```python
iris.drop([1,2,3,4]).head()
```

```python
# axis=1로 열을 삭제한다
iris=sns.load_dataset("iris")
iris.drop("species",axis=1).head()
```

```python
iris.drop(["species","sepal_length"],axis=1).head()
```

```python
iris.drop(["species","sepal_length"],axis='columns').head()
```

```python
#열 삭제와 재할당
iris.drop("species",axis=1).head()
```

```python
# 삭제해도 원본 데이터는 삭제 도지 않는다
print(iris.head())
# inplace=True 를 하면 삭제가 가능하다
iris.drop("species",axis=1,inplace=True)
print(iris.head())
```
