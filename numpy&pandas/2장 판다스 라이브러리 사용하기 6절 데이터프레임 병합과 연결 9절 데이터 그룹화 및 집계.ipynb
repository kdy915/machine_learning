### 데이터프레임 병합과 연결

```python
#merge()를 이용한 데이터프레임 병합
#동일한 키가 있는 값만 merge된다
import pandas as pd
df1=pd.DataFrame({'key':['a','b','c','f'],'c1':[1,2,3,5]})
df2=pd.DataFrame({'key':['a','b','c','d'],'c2':[5,6,7,8]})
df1.merge(df2)
```

```python
# how를 통해서 left,right,outer를 통해 데이터정보를 모두 합칠 수 있다.
df1.merge(df2,how='left')
```

```python
df1.merge(df2,how='right')
```

```python
df1.merge(df2,how='outer')
```

```python
#기준 열이 다를 경우
df3=pd.DataFrame({'key3':['a','b','c','f'],'c1':[1,2,3,5]})
df4=pd.DataFrame({'key4':['a','b','d','f'],'c2':[5,6,7,8]})
df3.merge(df4,left_on='key3',right_on='key4')
```

```python
#concat()을 이용한 데이터프레임 연결
df1=pd.DataFrame({'c1':[1,2,3,4],
                  'c2':[5,6,7,8]})
df2=pd.DataFrame({'c3':['a','b','c','d'],
                  'c4':[1.2,3.4,5.5,7.6]})
```

```python
pd.concat([df1,df2],axis=1,sort=False)
```

```python
df1=pd.DataFrame({"c1":[1,2],'c2':[3,4]})
df2=pd.DataFrame({"c1":[5,6],'c4':[7,8]})
pd.concat([df1,df2],axis=0,sort=False)
```

```python
df1=pd.DataFrame({'c1':[1,2,3,4],
                  'c2':[5,6,7,8]},index=[0,2,4,6])
df2=pd.DataFrame({'c3':['a','b','c','d'],
                  'c4':[1.2,3.4,5.5,7.6]},index=[0,1,2,3])
pd.concat([df1,df2],axis=1)
```

```python
#인덱스를 초기화 해서 값을 합치려면
df1.reset_index(drop=True,inplace=True)
pd.concat([df1,df2],axis=1)
```

### 정렬

```python
#열로 정렬하기
import seaborn as sns
iris=sns.load_dataset('iris')
iris.sort_index(axis=1,inplace=True)
iris
```

```python
#값으로 정렬
iris=sns.load_dataset("iris")
iris.sort_values(by="sepal_length",inplace=True)
iris.head()
```

```python
iris=sns.load_dataset("iris")
iris.sort_values(by=['sepal_length','sepal_width'],inplace=True)
iris.head()
```

```python
#레벨로 정렬
iris=sns.load_dataset('iris')
iris.columns=[['sepal','sepal','petal','petal','species'],iris.columns]
iris.columns.names=['info','detail']
iris.head()
```

```python
#레벨 이름으로 정렬하기
iris.sort_index(level=['detail'],ascending=False,axis=1).head()
```

```python
#레벨 이름으로 정렬하기
iris.sort_index(level=['info'],axis=1).head()
```
