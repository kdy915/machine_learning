## 혼자 풀이

```python
import pandas as pd
product = pd.read_csv('구매상품TR-D.csv',encoding='EUC-KR')
product.head()
```

```python
product_cls = pd.read_csv('상품분류-D.txt',encoding='EUC-KR')
product_cls.head()
```

```python
purchase_df=product.merge(product_cls, how='left')
```

```python
purchase_df['value']=1
purchase_df
```

```python
purchase_pivot = purchase_df.pivot_table(index='영수증번호',columns='소분류명',values='value',fill_value=0)
purchase_pivot
```

```python
purchase_pivot.shape[0]
```

```python
def transaction_list(df):
    total_list=[]
    items=df.columns
    count = df.shape[0]
    for j in range(count):
        result=[]
        for i, value in enumerate(df.iloc[j,:]):
            if value==1:
                result.append(items[i])
        total_list.append(result)
    return total_list
```

```python
tot_list = transaction_list(purchase_pivot)
tot_list
```

```python
from apyori import apriori
rules = apriori(tot_list,min_support=0.001, min_confidence=0.001)
results = list(rules)
```

```python
import pandas as pd
result_df = pd.DataFrame(None, columns=["LHS", "RHS", "support", "confidence", "lift"])
index=0

for row in results:
    support = row[1]
    ordered_stat = row[2]
    for ordered_item in ordered_stat:
        lhs = " ".join(x.strip() for x in ordered_item[0])
        rhs = " ".join(x.strip() for x in ordered_item[1])
        confidence = ordered_item[2]
        lift = ordered_item[3]
        result_df.loc[index] = [lhs, rhs, support, confidence, lift]
        index += 1
```

```python
result_df.sort_values(by=['support'],ascending=False)
```

```python
result_df.sort_values(by=['confidence'],ascending=False)
```

```python
result_df.sort_values(by=['lift'],ascending=False)
```
