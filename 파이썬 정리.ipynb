#클래스(class)

```python
class Person:
  name = 'kildong'

  def __init__(self, a:int, b:int=10) -> None:   #객체가 가져야할 값(a,b)
    self.a = a
    self.b = b
  
  def __del__(self) -> None:
    print('객체 소멸')

  def __str__(self):
    return f"a:{self.a}, b:{self.b}"

  def do_it(self, data) :
    print(data, self.a, self.b)
  
  @classmethod
  def do_that(cls):
    print(cls.name)
    
  @staticmethod
  def do_some():
    print('Hello')
```

```python
p=Person(10,20)
print(p)
```

```python
p=None
```

```python
p=Person(10,20)
del(p)
```

```python
p=Person(20,20)
p.do_it(100)
```

```python
Person.do_that() #class에 있는 class(name)을 직접 가져온다 클래스를 참조
```

```python
Person.do_some() #staticmethod 객체 생성하지않고 클래스 명과 함수를 불러 사용가능
```

```python
def text_preprocessing(text, delimiter=",.\'\"\n", isLower=True):
  if isLower:
    text = text.lower()
  for i in range(len(delimiter)):
    text=text.replace(delimiter[i],'')
  return text
```

```python
news_list = []
news=[news_1,news_2,news_3]
for text in news:
  news_list.append(list(set(text_preprocessing(text).split())))
```

```python
import json
with open('news.json','w',encoding='UTF-8-SIG') as f:
  json.dump(news_list,f)
```
